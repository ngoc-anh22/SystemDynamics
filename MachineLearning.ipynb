```python
data = pd.read_csv('Income.csv')
new_data = data[['construction','industry','transport', 'yearsmarried']]
new_data
```

In ML, 3 steps:

1- Create Model
2- Fitting data into the model
3- Prediction after training

```python
kmn = KMeans(n_clusters=3) # creat model : Clustring in this code we supposed 3 clusters
kmn.fit(new_data) # fitting data, the input of fitting can be a dataframe or array - all input should be numerical, can't be string
# the input can be continuous or distinst variable
Clusters = kmn.predict(new_data)
```

```python
Clusters
```

```python
df = pd.DataFrame(new_data, columns=['age','religiousness','affairs', 'yearsmarried'])
df['Clusters'] = Clusters
df
```

```python
xs = new_data['age']
ys = new_data['affairs']
labels = Clusters
centroids = kmn.cluster_centers_
plt.scatter(xs,ys,c=labels)
plt.xlabel('Age')
plt.ylabel('Affairs')
plt.scatter(centroids[:,0],centroids[:,2],marker='x', linewidth=15, s=25,c='red')
plt.show()
```

```python
new_data.data=new_data.values
```

```python
new_data.target=Clusters
```

```python
from sklearn.neighbors import KNeighborsClassifier # First step: creat model
knn = KNeighborsClassifier(n_neighbors=3, metric='minkowski',p=2) # creat model p是平方及平方根
x = new_data.data
y = new_data.target
knn.fit(x, y) # Second step:Fitting the model
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(new_data.data, new_data.target, test_size=0.3,stratify=new_data.target,random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train) # fitting the model based on the train samples
y_predict = knn.predict(x_test)
y_predict
```

```python
knn.score(x_test, y_test)
```

```python
neighbors = np.arange(1, 30)

train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
  knn_model = KNeighborsClassifier(n_neighbors=k)
  knn_model.fit(x_train, y_train)
  train_accuracy[i] = knn_model.score(x_train, y_train)
  test_accuracy[i] = knn_model.score(x_test, y_test)

plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth=6, min_samples_leaf=4 ) # Creat Model
dtc.fit(x_train, y_train) # Fit the model
```
