```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3,stratify=iris.target,random_state=42)
```

# **test_size**: This argument specifies the proportion of the dataset that should be allocated to the testing set. It can be a float value between 0 and 1, indicating the fraction of the dataset. For example, test_size=0.2 means that 20% of the data will be used for testing, and 80% will be used for training.

# **random_state**: This argument allows you to set a random seed for reproducibility. When you use the same random_state value, you ensure that the data split remains the same across different runs of your code. It's important for consistent results and debugging.Using a specific value for random_state ensures that you get the same split every time you run your code, which can be useful for debugging and sharing results.

# **stratify**: This argument is used for stratified sampling, which ensures that the class distribution in the original dataset is preserved in both the training and testing sets. It's often used when dealing with imbalanced datasets.

# **Stratify** : the propertional of target  is in equal train and test

# This stratify parameter makes a split so that the proportion of values in the sample produced will be the same as the proportion of values provided to parameter stratify.
# For example, if variable y is a binary categorical variable with values 0 and 1 and there are 25% of zeros and 75% of ones, stratify=y will make sure that your random split has 25% of 0's and 75% of 1's.

# **We use Train Sample for fitting data and use Test Sample for prediction **

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train) # fitting the model based on the train samples
y_predict = knn.predict(x_test)
y_predict
```

```python
knn.score(x_test, y_test)
```

### **Over Fitting and Under Fitting**

Over fitting means the accuracy of model for train sample is very nice and perfect near to 1.0 but for test sample the perfomace ( accuracy) is very low which shows that we can not generalize the model for different samples.

 Under Fitting means that the model is too simple to classify the objects correctely and we can not genralize the model to all samples.

 The best situation is the same accuracy and high enough ( for example around 70 to 95 % accuracy) between train and test samples.  

```python
neighbors = np.arange(1, 30)

train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
  knn_model = KNeighborsClassifier(n_neighbors=k)
  knn_model.fit(x_train, y_train)
  train_accuracy[i] = knn_model.score(x_train, y_train)
  test_accuracy[i] = knn_model.score(x_test, y_test)

plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth=6, min_samples_leaf=4 ) # Creat Model
dtc.fit(x_train, y_train) # Fit the model
```

**max_depth**: Restrict the maximum depth of the tree to prevent the model from becoming too complex and overfitting.
**min_samples_split**: Set the minimum number of samples required to split an internal node. Increasing this value can result in a less complex model.
**min_samples_leaf**: Specify the minimum number of samples a leaf node must have. Larger numbers prevent the model from fitting excessively to the training data.
**max_features**: Limit the number of features to consider when looking for the best split; using fewer features can reduce overfitting.
**max_leaf_nodes**: Restrict the number of leaf nodes in the tree to simplify the model.

There are different types of nodes in a decision tree:

Root Node:

The topmost node of the tree.
Represents the entire dataset, which is then split based on the feature that provides the best split according to the chosen criterion (e.g., Gini impurity, entropy).
Internal Nodes:

Nodes that split into two or more sub-nodes.
Each internal node represents a decision based on one of the input features.
The decision at an internal node determines which subsequent node (or sub-node) a data point moves to.
Leaf Nodes:

The terminal nodes of the tree, also known as terminal or end nodes.
They do not split further.
Each leaf node represents a class label (in classification) or a continuous value (in regression).
